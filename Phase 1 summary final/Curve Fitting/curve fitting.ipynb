```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
from scipy.optimize import curve_fit
from scipy import interpolate
```

```python
df = pd.read_csv('Amreli NRM data for model building final.csv')


df = df[df['Heat number (life number)']>1]    
df = df.reset_index(drop=True)
df['Number of capacitors removed'].fillna(0, inplace=True)

threshold = 150

for i in range(df.shape[0]):
    if(df.loc[i,'Total Heat Time previous (minutes)']>threshold):
        df.loc[i,'Total Heat Time previous (minutes)']= threshold
    if(df.loc[i,'Total Heat Time (minutes)']>threshold):
        df.loc[i,'Total Heat Time (minutes)']= threshold
        
df['Input energy previous (KWH)'] = df['Power previous']*df['Total Heat Time previous (minutes)']/60
df['Input energy (KWH)'] = df['Power']*df['Total Heat Time (minutes)']/60
```

# Exponential curve fitting for Frequency

```python
df1 = df[df['campaign']==1]
y = df1['Frequency']
x = [i for i in range(df1.shape[0])]

def func(x, a, b):
    return (a*(np.exp(b*x)))
```

```python
param, param_cov = curve_fit(func, x, y)
```

```python
t = []
for i in x:
    t.append(func(i, param[0], param[1]))

plt.plot(y,c='g')
plt.xlabel('Heat Number')
plt.ylabel('Frequency')
plt.title('Actual data')
plt.show()
plt.plot(t, c='r')
plt.xlabel('Heat Number')
plt.ylabel('Frequency')
plt.title('Exponential fitting');
plt.show()


plt.plot(y,c='g')
plt.plot(t, c='r')
plt.xlabel('Heat Number')
plt.ylabel('Frequency')
plt.title('Actual data')
plt.legend(['Actual data', 'Curve Fitting data']);
```

# Quadratic fitting for lining

```python
df1 = df[df['campaign']==1]
y = df1['Refractory lining before heating']
x = [i for i in range(df1.shape[0])]

def func(x, a, b):
    return ((a)-(b*x*x))
```

```python
l1 = (y[len(x)-1])-(y[0])
l2 = ((len(x)-1)*(len(x)-1))
d = -l1/l2
c = y[0]

yn = []
for i in x:
    yn.append(func(i, c, d))
```

```python
plt.plot(y)
plt.plot(yn)
plt.xlabel('Heat number')
plt.ylabel('Lining thickness')
plt.legend(['Old Refractory lining', 'Proposed Refractory lining'])
```

# Creating dataset with quadratic lining

```python
df = pd.read_csv('Amreli NRM data for model building final.csv')
```

```python
lining = []
campaigns = np.unique(df['campaign'])

def func(x, a, b):
    return ((a)-(b*x*x))

for c in campaigns:
    dfc = df[df['campaign']==c].reset_index(drop=True)
    #y = dfc['Refractory lining before heating']
    x = [i for i in range(dfc.shape[0]+1)]
    l1 = -(6.75)+(3.916667)
    l2 = ((len(x)-1)*(len(x)-1))
    d = -l1/l2
    c = 6.75
    for i in range(len(x)-1):
        lining.append(func(x[i], c, d))
```

```python
df['Refractory lining before heating'] = lining
```

```python
lining1 = []
campaigns = np.unique(df['campaign'])

def func(x, a, b):
    return ((a)-(b*x*x))

for c in campaigns:
    dfc = df[df['campaign']==c].reset_index(drop=True)
    
    x = [i for i in range(dfc.shape[0]+1)]
    l1 = -(6.75)+(3.916667)
    l2 = ((len(x)-1)*(len(x)-1))
    d = -l1/l2
    c = 6.75
    for i in range(1, len(x)):
        lining1.append(func(x[i], c, d))
```

```python
df['Refractory lining after heating'] = lining1
```

```python
df.to_csv('Amrel NRM data for model building final with quadratic lining.csv', index=False)
```
